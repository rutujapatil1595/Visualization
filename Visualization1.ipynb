```python
#Scatter plot
plt.figure(figsize=(5,5))
plt.title("Width v/s price of car")
x=car['width']
y=car['price']
plt.scatter(x,y)
plt.xlabel('width of car')
plt.ylabel('price of car')
plt.show()
```

```python
#Histogram
plt.hist(car['fuel-type'])
plt.xlabel('Fuel Type')
plt.ylabel('No of cars')
plt.title("Fuel type vs no of cars")
c=car['fuel-type'].value_counts()
plt.show()
```

```python
#Bar plot
car['fuel-type'].value_counts().plot(kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('No of cars')
plt.title("Fuel type vs no of cars")
c=car['fuel-type'].value_counts()
plt.show()
```

```python
#pie plot
car['body-style'].value_counts().plot(kind='pie',autopct='%.2f%%',explode=(0,0.3,0,0.5,0))
plt.show()
```

```python
#Countplot
plt.figure(figsize=(5,5))
sns.countplot(data=car,x='body-style')
f=car['body-style'].value_counts()
plt.yticks(f)
plt.show()
```

```python
#Boxplot
X=car.iloc[:,:-1] #retrive all col but  except price
#X=car['syboling','city-mpg','highway-mpg'......]
target=car.iloc[:,-1]  #price
#target=car['price']
sns.boxplot(data=X,x=target,y='make')
plt.grid()
```
