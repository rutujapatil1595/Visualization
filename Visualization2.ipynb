```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
area=np.array([600,800,1000,1200,1500,2000])
price=np.array([4,6,9,12,16,21])           #lakhs
print(area)
print(price)
```

```python
#visualizing above data using scatter plot
plt.scatter(area,price,marker='*',c='green',s=50)
plt.title("Area vs Price")
plt.xlabel('Area in sqft',c='blue')
plt.ylabel('Price in Lakhs',c='blue')
plt.show()
```

```python
#line plot
plt.plot(area,price,c='red')
plt.scatter(area,price,c='green',marker='*',s=50)
plt.title("Area vs Price")
plt.xlabel('Area in sqft',c='blue')
plt.ylabel('Price in Lakhs',c='blue')
plt.show()
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic['who'].value_counts()
```

```python
titanic.info()
```

```python
#count plot on class col
sns.countplot(data=titanic,x='class')
f=titanic['class'].value_counts()
plt.yticks(f)        #indicate exact count of people in each class
plt.show()
```

```python
#count plot on who col
sns.countplot(data=titanic,x='class',hue='who')
plt.show()
```

```python
#count plot
sns.countplot(data=titanic,x='survived',hue='pclass')
plt.show()
```

```python
#pie chart
titanic['pclass'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.show()
```
